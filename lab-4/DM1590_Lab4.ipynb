# 1.5* 

From the results in 1.3, identify the four pairs of dimensions that have the largest magnitude correlation and make a scatter plot matrix of them.

```python
import numpy as np

corr_abs = corr_matrix.abs()

mask = np.triu(np.ones_like(corr_abs, dtype=bool), k=1) #upper triangle to remove 1s and symmetry
corr_values = corr_abs.where(mask)

top4 = corr_values.unstack().dropna().sort_values(ascending=False).head(4) #top 4 correlations

features = list(set([f for pair in top4.index for f in pair])) #what features the top 4 has

sns.pairplot(full_data[features])
plt.suptitle("Scatter Plot Matrix of Top Correlated Features")
plt.show()
```

# 1.7* 

Compute the covariance matrix $\mathbf{C}$ of the dimensions in the dataset, and perform an eigenvector decomposition of it, $\mathbf{C} = \mathbf{Q}\mathbf{\Lambda}\mathbf{Q}^T$. Plot the eigenvalues, i.e., the diagonal elements of $\mathbf{\Lambda}$. 

```python
corr_matrix = np.corrcoef(full_data, rowvar=False)
eigenvalues, eigenvectors = np.linalg.eigh(corr_matrix)
plt.plot(eigenvalues, "o-")
plt.show()
```

# 1.8*

Plot the two eigenvectors in $\mathbf{Q}$ having the largest eigenvalues, and label the x-axis with the appropriate dimensions.


```python
# add your code below
indices = np.argpartition(eigenvalues, -2)[-2:]
# To get them in exact descending order of value:
indices = indices[np.argsort(eigenvalues[indices])][::-1]
for index in indices:
    plt.plot(eigenvectors[index])
plt.show()
```
