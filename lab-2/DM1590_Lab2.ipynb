# 1.2*
Compute the Euclidean and Manhattan distances between the two points.

```python
def euclid(a, b):
    return np.sqrt( ((a[0] - b[0])**2) + ((a[1] - b[1])**2) + ((a[2] - b[2])**2) )

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1]) + abs(a[2] - b[2])

print(euclid(noa, erik))
print(manhattan(noa, erik))

```

# 1.3*

Find the point that is half-way between the two points and plot it on the 3D plot you created in 1.1. What is the zodiac sign of the person who is described by that midway point?

```python
def halfway(a, b):
    return (a + b) / 2

noa = np.array([22, 181, 5])
erk = np.array([22, 194, 5])
halfway_point = halfway(noa, erk)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(*noa, 'rd')
ax.plot(*erk, 'rd')
ax.plot(*halfway_point, 'bo')

ax.set_xlabel('Age')
ax.set_ylabel('Height')
ax.set_zlabel('Zodiac')

plt.show()
#print(halfway_point)
print(f"Halfway point is: {np.floor(halfway_point[2])}")
```

**Answer**: The zodiac sign is coded 5, which is *Cancer*!

# 1.4*

Now create 1000 two-dimensional vectors where each element is sampled from a normal distribution with mean 0 and standard deviation 1. Make a scatterplot of these points.

```python
scatter_x = np.random.normal(loc=0.0, scale=1.0, size=1000)
scatter_y = np.random.normal(loc=0.0, scale=1.0, size=1000)
plt.scatter(scatter_x, scatter_y)
plt.show()
```

# 1.5* 

Multiply each one of these random vectors by the matrix [2 1; 1 1] and plot the resulting scatterplot.

```python
combined = np.column_stack((scatter_x, scatter_y))
matrix = np.array([[2, 1],[1, 1]])
dot_combined = np.dot(combined, matrix)
plt.scatter(dot_combined[:,0], dot_combined[:,1])
plt.show()
```

# 1.8* 

Now create 1000, 2-dimensional points with each dimension uniformly distributed in [-1,1]. For each of these points, find the Euclidean distance to all other points. Make a histogram of the distribution of these distances. Set the limit of the xaxis to be between 0 and 10.

```python
uniform = np.random.uniform(low=-1, high=1, size=[1000, 2])

distance = []
for i in range(1000):
    for j in range(1000):
        distance.append(np.linalg.norm(uniform[i, :] - uniform[j, :]))

plt.xlim(0, 10)
plt.hist(distance, bins = 40)
plt.show()
```
