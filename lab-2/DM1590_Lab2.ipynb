Build two vectors that numerically describe you and your lab partner.
```python
# see mpl_toolkits.mplot3d, in particular Axes3D
# add your code below
import matplotlib.pyplot as plt
import numpy as np

noa = np.array([22, 181, 3])
erk = np.array([22, 194, 3])

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(*noa, 'rd')
ax.plot(*erk, 'bo')

ax.set_xlabel('Age')
ax.set_ylabel('Height')
ax.set_zlabel('Zodiac')

plt.show()

```

# 1.2*
Compute the Euclidean and Manhattan distances between the two points.

```python
# add your code below

def euclid(a, b):
    return np.sqrt( (a[0] - b[0])**2 + (a[1] - b[1])**2 + (a[2] - b[2])**2 )

def manhattan(a, b):
    return abs(a[0] - b[0]) + abs(a[1] - b[1]) + abs(a[2] - b[2])

print(euclid(noa, erk))
print(manhattan(noa, erk))

```

# 1.3*

Find the point that is half-way between the two points and plot it on the 3D plot you created in 1.1. What is the zodiac sign of the person who is described by that midway point?

```python
# add your code below

def halfway(a, b):
    return (a + b) / 2

noa = np.array([22, 181, 3])
erk = np.array([22, 194, 3])
halfway_point = halfway(noa, erk)

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

ax.plot(*noa, 'rd')
ax.plot(*erk, 'rd')
ax.plot(*halfway_point, 'bo')

ax.set_xlabel('Age')
ax.set_ylabel('Height')
ax.set_zlabel('Zodiac')

plt.show()
#print(halfway_point)
print(np.floor(halfway_point[2])) # => Halfway is
```
